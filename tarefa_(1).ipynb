```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lendo os dados

notas=pd.read_csv('https://github.com/danieltb3006/DM_Ibmec/raw/main/notas_alunos_Ibmec.csv',encoding='ISO-8859-1',sep=';',decimal='.')
notas
```

```python
# Removendo dados faltantes

notas=notas.dropna()
notas
```

```python
# 1) Inserir uma coluna e calcular a média seguindo as regras de pesos do Ibmec (considere AP3 como AC)

notas['Media']=notas['AP1']*0.4+notas['AP2']*0.4+notas['Ap3']*0.2

notas
```

```python
# 2) Inserir uma coluna informando a situação final do aluno (aprovado, reprovado ou substitutiva)
# Aprovado: media>7
# Substitutiva: 3 <= media <=7
# reprovado: media<3

notas['Situacao']=np.where(notas['Media']>=7,'Aprovado',np.where(notas['Media']>=3,'Substitutiva','Reprovado'))

notas
```

```python
notas["Curso"].unique()
```

```python
# 3) Escrever por extenso os nomes dos cursos

cursos = {
    'Adm': 'Administração',
    'Dir': 'Direito',
    'Eco': 'Economia',
    'RI': 'Relações Internacionais',}
notas['Curso por extenso'] = np.array([cursos[i ] for i in notas['Curso']])

notas
```

```python
# 4) Montar a fórmula para responder as seguintes perguntas:

# - Média dos alunos por cursos
# - Menor média dos alunos por cursos
# - Maior média dos alunos por cursos

# - Média dos alunos por cursos

media_por_curso = notas.groupby('Curso por extenso')['Media'].mean()

for curso in media_por_curso.index:
    print(f'Média do curso {curso}: {media_por_curso[curso]:.2f}')

# - Maior média dos alunos por cursos

notas.groupby(by='Curso').max()['Media']

```

```python
# - Menor média dos alunos por cursos

notas.groupby(by='Curso').min()['Media']
```

```python
# 5) Neste exercício é obrigatório apresentar aplicações dos seguintes comandos:
#  - tabela de frequência (value_counts)
#  - tabela usando agrupamentos (groupby)
#  - tabela usando tabela dinâmica (pivot_table)
#  - tabela usando contagem (crosstab)
#  - subtabelas com critérios (query)

# - tabela de frequência (value_counts)

notas['Curso'].value_counts()
```

```python
# - tabela usando agrupamentos (groupby)

notas.groupby('Curso')['Media'].mean()
```

```python
# - tabela usando tabela dinâmica (pivot_table)

pd.pivot_table(notas, index='Curso', values='Media', aggfunc='mean')
```

```python
# - tabela usando contagem (crosstab)

pd.crosstab(notas['Curso'], notas['Situacao'])
```

```python
# 6) Gerar gráficos (1 de cada): barras, colunas, histograma, boxplot

# - barras

notas['Curso'].value_counts().plot.bar()
```

```python
# - colunas

notas['Curso'].value_counts().plot.barh()
```

```python
# - histogramm 

notas['Media'].plot.hist()
```

```python
# - boxplot

notas.boxplot(column='Media', by='Curso')
```
